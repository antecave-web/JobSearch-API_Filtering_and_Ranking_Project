```python
import requests
import time

SEARCH_TERMS = [
    "utredare",
    "handläggare",
    "analytiker",
    "dataanalytiker",
    "data analyst",
    "data analytiker",
    "business analyst",
    "research assistant",
    "forskningsassistent",
    "utvärderare",
    "verksamhetsutvecklare",
    "kvalitetsutvecklare",
    "controller",
    "statistiker",
    "policy analyst",
    "administratör statistik",
]
#SEARCH_TERMS = ["utredare"] # för debugging
    
SEARCH_MODS = [
    "Stockholm",
    "distans",
    "hybrid",
    "remote",
]
EXCLUDE = [
    "senior",
    "lead",
    "principal",
    "staff engineer",
    "architect",
    "10 years",
    "8 years",
    "specialist",
    "expert",
    "head of",
    "manager",
]
BOOST = [
    "analys",
    "utredning",
    "rapport",
    "uppföljning",
    "myndighet",
    "offentlig",
    "samhäll",
    "kvalitet",
    "verksamhetsutveckling",
    "excel",
    "sql",
    "power bi",
    "statistik",
    "samordning",
    "administration",
    "utvärdering",
]

BASE_URL = "https://links.api.jobtechdev.se"

all_jobs = []
seen_ids = set()

for term in SEARCH_TERMS:
    for mod in SEARCH_MODS:

        query = f"{term} {mod}"

        params = {
            "q": query,
            #"limit": 20,
            "sort": "pubdate-desc"
        }

        print(f"Searching: {query}")

        response = requests.get(
            f"{BASE_URL}/joblinks",
            params=params
        )

        time.sleep(0.5) #pause in between each api requests

        print("Status:", response.status_code)

        if response.status_code != 200:
            continue

        data = response.json()

        for job in data.get("hits", []):

            # Adjust this if API uses another ID field
            job_id = job.get("id")

            # Skip duplicates
            if job_id in seen_ids:
                continue

            seen_ids.add(job_id)

            all_jobs.append(job)

print(f"\nUnique jobs found: {len(all_jobs)}")

# Print results
for job in all_jobs:

    title = job.get("headline", "No title")
    employer = job.get("employer", "Unknown employer")
    url = job.get("source_links", "No url")[0]['url']

    print("\n-------------------")
    print("Title:", title)
    print("Employer:", employer)
    print(url)
```

```python
import sqlite3 #save jobs to SQLite database
import json

conn = sqlite3.connect("jobs.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS jobs (
    id TEXT PRIMARY KEY,
    raw_json TEXT
)
""")

for job in all_jobs:

    cursor.execute("""
    INSERT OR IGNORE INTO jobs
    VALUES (?, ?)
    """, (
        job.get("id"),
        json.dumps(job, ensure_ascii=False)
    ))

conn.commit()
conn.close()
```
